G) CPM (critical path method) in Python

```python
def topological_sort(graph):
    n = len(graph)
    visited = [False] * n
    stack = []

    def dfs(v):
        visited[v] = True
        for i, weight in enumerate(graph[v]):
            if weight and not visited[i]:
                dfs(i)
        stack.insert(0, v)

    for i in range(n):
        if not visited[i]:
            dfs(i)

    return stack

def longest_path_to_node(graph, target_node):
    order = topological_sort(graph)
    distances = [-float('inf')] * len(graph)
    predecessor = [None] * len(graph)
    start_node = order[0]
    distances[start_node] = 0

    for node in order:
        for adj in range(len(graph[node])):
            if graph[node][adj] != 0 and distances[adj] < distances[node] + graph[node][adj]:
                distances[adj] = distances[node] + graph[node][adj]
                predecessor[adj] = node

    path = []
    current = target_node
    while current is not None:
        path.insert(0, current)
        current = predecessor[current]

    if path[0] != start_node:
        return None

    return path, distances[target_node]


critical_path, path_length = longest_path_to_node(w, 11)
if critical_path:
    print("Critical path to node 11:", critical_path)
    print("Length of the critical path:", path_length)
else:
    print("No path to node 11.")
```

Process Overview:

The process involves the financial dynamics of a company’s revenue before and after strategic changes such as investments in advertising. This model demonstrates how these strategic decisions can significantly affect the company's financial trajectory.

image.png

Functionality Before Changes:

I(t) = 1000000*(1+0,1)**t-500000*(1+0,05)**t

This function represents the company’s revenue growth, which increases linearly over time, offset by quadratic growth in costs associated with maintaining this growth.



Functionality After Changes:

I(t) = 1000000*(1+0,2)**t-500000*(1+0,1)**t

After investing in advertising, revenue increases more steeply due to enhanced exposure and customer reach, albeit with higher associated costs, as reflected in the steeper quadratic term.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions to represent the company's revenue before and after strategic changes
def initial_income(t):
    return 1000000*(1+10/100)**t-500000*(1+5/100)**t

def modified_income(t):
    return 1000000*(1+20/100)**t-500000*(1+20/100)**t

# Define the time array to represent years (0 to 5 years)
t = np.linspace(0, 5, 100)

# Calculate revenue values using the defined functions
initial_values = initial_income(t)
modified_values = modified_income(t)

# Calculate the area under the revenue curves (integral approximation)
area_initial = np.trapz(initial_values, t)
area_modified = np.trapz(modified_values, t)
cost_of_decision = area_modified - area_initial
print(cost_of_decision)

# Generate plots to visualize the revenue functions
plt.figure(figsize=(14, 8))
plt.plot(t, initial_values, label='Initial Income', color='red')
plt.plot(t, modified_values, label='Modified Income', color='orange')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# A function that represents the company's revenues before and after strategic changes
def initial_income(t):
    return 1000000*(1+10/100)**t-500000*(1+5/100)**t

def modified_income(t):
    return 1000000*(1+20/100)**t-500000*(1+20/100)**t

# Calculate derivatives
def derivative(f, a, h=0.01):
    """ The derivative of f at point a """
    return (f(a + h) - f(a - h)) / (2 * h)

# Time (years)
t = np.linspace(0, 5, 100)

# Calculating function values and derivatives
initial_values = initial_income(t)
modified_values = modified_income(t)
initial_derivative = [derivative(initial_income, i) for i in t]
modified_derivative = [derivative(modified_income, i) for i in t]

# Drawing charts
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Graph of the income function
ax1.plot(t, initial_values, label='Initial Income')
ax1.plot(t, modified_values, label='Modified Income')
ax1.set_title('Income Over Time')
ax1.set_xlabel('Time (years)')
ax1.set_ylabel('Income')
ax1.legend()

# Income derivative chart
ax2.plot(t, initial_derivative, label='Derivative of Initial Income')
ax2.plot(t, modified_derivative, label='Derivative of Modified Income')
ax2.set_title('Rate of Change of Income Over Time')
ax2.set_xlabel('Time (years)')
ax2.set_ylabel('Rate of Change of Income')
ax2.legend()

plt.tight_layout()
plt.show()
```

The design office is working on introducing a new line of men's suits and accessories dedicated to each set, such as footwear, watches and jewelry. The projects require extensive knowledge and commitment of employees, and work schedules overlap. The goal is to maximize productivity by optimizing employees' working time on individual project stages. For customer satisfaction, it is also necessary for work to be carried out on time without exceeding the total number of man-hours available.

```python
import numpy as np
from scipy.optimize import linprog

# Productivity metrics for each project and employee,
# For minimization, the coefficients are negative
C = [-2, -6, -4, -1, -2, -6]

# Coefficients of the inequality constraints
A = [
    [1, 1, 0, 0, 0, 0],  # Max hours constraint for Employee 1
    [0, 0, 1, 1, 0, 0],  # Max hours constraint for Employee 2
    [0, 0, 0, 0, 1, 1],  # Max hours constraint for Employee 3
    [-1, 0, -1, 0, -1, 0],  # Max hours constraint for Project 1
    [0, -1, 0, -1, 0, -1]   # # Max hours constraint for Project 2
]

# Right-hand side of the inequality constraints and adjusted project hours for feasibility
b = [40, 40, 40, -60, -60]

# Bounds for each variable (hours must be non-negative)
bounds = [(0, None) for _ in range(6)]

# Linear programming problem and solving
result = linprog(C, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# The result
if result.success:
    print("Optimal hours distribution among employees and projects:")
    print("Employee 1, Project 1:", result.x[0])
    print("Employee 1, Project 2:", result.x[1])
    print("Employee 2, Project 1:", result.x[2])
    print("Employee 2, Project 2:", result.x[3])
    print("Employee 3, Project 1:", result.x[4])
    print("Employee 3, Project 2:", result.x[5])
    print(f"Maximized productivity: ${-result.fun}")
else:
    print("Optimization has been failed:", result.message)
```
